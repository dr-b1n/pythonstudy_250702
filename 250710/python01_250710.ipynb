#### 파이썬 크롤링 미션
- https://crawlingtest.netlify.app/ 해당 사이트에서 course라는 클래스값을 갖고 있는 li태그 내부 텍스트를 출력하세요
- 단, 왕초보 혹은 초급, - 와 같은 부가적인 설명을 제외하고, 실제 해당 과정의 설명문안만 크롤링 해보세요!

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/")
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find_all("li", attrs = {"class":"course"})

for titles in section :
    text = titles.get_text()
    # print(text)
    # 리스트들을 for문으로 펼쳐서 프린트 해주고 싶음
    for texts in text.split(" - ") :
        print(texts.strip("왕초보"))
```

```python
# 정답

import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/")
soup = BeautifulSoup(res.content, "html.parser")

titles = soup.find_all("li", "course")

for title in titles :
    print(title.get_text().split("-")[-1].replace("\n", " ").replace(" ", "").strip())
```
